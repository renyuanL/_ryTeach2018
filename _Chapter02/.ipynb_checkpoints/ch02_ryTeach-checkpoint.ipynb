```python
import pandas as pd

df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', 
                 header= None)
```

```python
#%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

Labels=   (-1, +1)
Markers=  ('x',   'o')
Colors=   ('blue','red')

#
# 數據預處理
#
# 從 df[4] (代表類別) 選 'Iris-setosa' 或 'Iris-versicolor'
# df ==> df2
#
df2= df[(df[4] == 'Iris-setosa') | 
        (df[4] == 'Iris-versicolor')]

print('df2= ', df2)
print('df2.describe()= ', df2.describe())

#
# 抓出特徵
# extract sepalLength@Column[0] and petalLength@Column[2]
X= df2[[0,2]].values
#print('X= ',X)

# 抓出類別
y= df2[4].values
#print('y= ',y)

# 文字類別 ('Iris-setosa', 非 'Iris-setosa')
# ==> 數字類別 (+1, -1)
y= np.where(y == 'Iris-setosa', # if
            -1,                 # then
            +1)                 # else
#print('y= ',y)

uniqueY= np.unique(y)
#print('uniqueY= ', uniqueY)

#
# 數據做圖
#

# 陽春版
x0= X[:, 0], 
x1= X[:, 1]
plt.scatter(x0, x1)
plt.xlabel('x0')
plt.ylabel('x1')
plt.show()

# 精細版
for i, b in enumerate(Labels): 
    # b= -1, +1
    x0= X[y == b, 0] 
    x1= X[y == b, 1]

    plt.scatter(x= x0, 
                y= x1, 
                label= b,
                c=      Colors[i],
                marker= Markers[i], 
                edgecolor='black')

plt.xlabel('x0')
plt.ylabel('x1')
plt.legend()
plt.show()
```

```python
import numpy as np

from matplotlib.colors import ListedColormap

markers=  ('x',   'o')
colors=   ('blue','red')
colormap= ListedColormap(colors)
    
def plot_decision_regions(X, y, classifier, resolution= .1):

    # setup marker generator and color map
     
    # 以本題而言，  '-1', '+1' ==> index= [0, 1] ==> ['x', 'o']
    uniqueY= np.unique(y) # == [-1, +1]
    # np.unique(y) Returns the sorted unique elements of an array.
    
    # plot the decision surface
    x0_min, x0_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x1_min, x1_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    
    x0_min= x1_min = 0
    
    #製造所有範圍內的「網格點」
    xx0, xx1= np.meshgrid(np.arange(x0_min, x0_max, resolution),
                          np.arange(x1_min, x1_max, resolution))
    
    # 全面對所有網格點做 predict (工程浩大喔！) 
    Z= classifier.predict(np.array([xx0.ravel(), xx1.ravel()]).T)
    
    Z= Z.reshape(xx0.shape)
    
    plt.contourf(xx0, xx1, Z, alpha= .1, cmap= colormap)
    #plt.pcolormesh(xx0, xx1, Z, alpha= .1, cmap= colormap)
    
    plt.xlim(xx0.min(), xx0.max())
    plt.ylim(xx1.min(), xx1.max())

    # plot class samples
    for idx, cl in enumerate(uniqueY):
        
        x0= X[y == cl, 0] 
        x1= X[y == cl, 1]
            
        plt.scatter(x= x0, 
                    y= x1, 
                    label= cl,
                    alpha= .9, 
                    c=      colors[idx],
                    marker= markers[idx], 
                    edgecolor='black')

class Perceptron: #(object):
    """Perceptron classifier.

    Parameters
    ------------
    eta : float
      Learning rate (between 0.0 and 1.0)
    n_iter : int
      Passes over the training dataset.
    random_state : int
      Random number generator seed for random weight
      initialization.

    Attributes
    -----------
    w_ : 1d-array
      Weights after fitting.
    errors_ : list
      Number of misclassifications (updates) in each epoch.

    """
    def __init__(self, eta= 0.01, n_iter= 100, random_state= 1):
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state

    def fit(self, X, y, decisionRegionDraw= False):
        """Fit training data.

        Parameters
        ----------
        X : {array-like}, shape = [n_samples, n_features]
          Training vectors, where n_samples is the number of samples and
          n_features is the number of features.
        y : array-like, shape = [n_samples]
          Target values.

        Returns
        -------
        self : object

        """
        rgen= np.random.RandomState(self.random_state)
        
        self.w_     = rgen.normal(loc=   -10.0, 
                                  scale= 0.1, 
                                  size=  1 + X.shape[1])
        self.errors_= []

        for i in range(self.n_iter):
            errors = 0
            for xi, yi in zip(X, y):
                
                ### Rosenblatt's Perceptron Learning Rule:
                
                 
                yHat= self.predict(xi) 
                e=    yi - yHat
                
                x=          np.append([1], xi)
                self.w_ +=  x * e * self.eta
                
                if (yi != yHat):
                    errors += 1 

            
            self.errors_.append(errors)
            
            ### ry: 每一次 fit 畫一下 decision region
            #'''
            if decisionRegionDraw == True:
                print('i= [{}], errors= {}, self.w_= {} '.format( i, errors, self.w_))
                plot_decision_regions(X, y, classifier= self) #ppn)
                plt.show()
            #'''
        ### ry
        #plt.show()
            
        return self

    def net_input(self, X):
        """Calculate net input"""
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def predict(self, X):
        """Return class label after unit step"""
        return np.where(self.net_input(X) >= 0.0, +1, -1)
```

```python
ppn= Perceptron(eta= 0.1, n_iter= 10)

ppn.fit(X, y, decisionRegionDraw= True)

print('ppn.w_= ',ppn.w_)

plt.plot(range(1, len(ppn.errors_) + 1), 
         ppn.errors_, 
         marker='o')
plt.xlabel('Epochs (時代，輪)')
plt.ylabel('Errors (錯誤數)')

# plt.savefig('images/02_07.png', dpi=300)
plt.show()
```
